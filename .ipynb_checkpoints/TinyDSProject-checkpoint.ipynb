#### At this point i encounter a big problem. One of the datasets, "unemployment analysis.csv" stores the data in a completely different format than the rest. This dataset includes information about the unemployment rate for a set of countries over some years. Through my pre-processing, the year range is from 2000-2018 which matches the rest of the datasets, and the countries are countries that are found in all the other datasets. However, this dataset stores the unemployment rate for each year in a seperate column, rather than having a 'year' column with the unemploymnet rate next to it as its own column. I need to reformat it in order to ba able to combine it with the other datasets. Thankfully, this can be done by using the .melt library from pandas.

```python
df_unemployment2 = df_unemployment1.melt(id_vars=["Country Name", "Country Code"], 
        var_name="Year", 
        value_name="Unemployment_rate")

print(df_unemployment2)
```

#### Now we brought it to a format that will allow us to merge the datsets. However, we must first sort this new formatted dataset base on "country name" and "year" so that records can be matched appropriately when the dataframes are combined.

```python
df_unemployment2 = df_unemployment2.sort_values(by=['Country Name', 'Year'])
print(df_unemployment2)
```

#### Now we can add the desired columns from this dataset into our final dataframe

```python
df_gapminder['Country_Code']= df_unemployment2['Country Code'].values
df_gapminder['Unemployment_Rate']= df_unemployment2['Unemployment_rate'].values
print(df_gapminder)
```

### There are a few missing values in our dataframe -- 8 in "hdi_index" column and 2 in "gdp" column. Since we have data for later years, we can infer these data through various methods.

### First lets tackle the "gdp" column since it only has 2 values missing. These values are for Afghanistan for the years 2000-2001. 

### To infer these 2 values we can calculate the mean from 2002-2003 (2 years) since we are missing 2 years and use that mean as the inferred value of our missing values.

## Alternative method:

### After considering the non missing values, we can see that gdp for Afghanistan was rising from 2002-2013 while it started to gradually fall the next years since i am interested in the years prior to 2002 (2000-2001), it would be a logical assumption that there was a rising trend in those years as well. Therefore, i decided to infer the values of those years by calculating the rate of increase from 2002-2013

### i.e. (346-333) + (336-346) + (360-336) + (368-360) + (409-368) + (415-409) + (492-415) + (547-492) + (533-547) + (580-533) + (592-580) = 259 => 259/11 = 23.545 => 24

### and then subtract this from the gdp value of a year to infer the previous year

```python
# Next step is to tackle the problem of missing values
# There are values missing  from our gapminder dataset, specifically from the "hdi index" and "gdp" columns

gdp_Afghan_infer = df_gapminder.loc[(df_gapminder['country'] == "Afghanistan") & (df_gapminder['year'] == 2002), 'gdp'].item() +  df_gapminder.loc[(df_gapminder['country'] == "Afghanistan") & (df_gapminder['year'] == 2003), 'gdp'].item()
gdp_Afghan_infer = gdp_Afghan_infer/2
print("mean gdp value for 2002-2003 is: ", gdp_Afghan_infer)

df_gapminder["gdp"].fillna(gdp_Afghan_infer, inplace = True)
#df_gapminder
```

### Now we can tackle the 8 missing values in the "hdi_index" column

### For this column we are missing values for Nigeria for the years 2000-2002 and for Lebanon for the years 2000-2004

### To fill this values i will use a pandas method, .fillna(), to infer the missing values from the values of the years that follow.

```python
df_gapminder["hdi_index"].fillna( method ='bfill', inplace = True)
```

### This pandas method finds the first non-missing value after the null value and backfills it to the missing data. In our case it took the value for the year 2003 for Nigeria and filled the years 2000-2002.

### In the case of Lebanon, it took the value for year 2005, and replaced the missing values of years 2000-2004 with that.
